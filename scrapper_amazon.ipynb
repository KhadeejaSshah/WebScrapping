```python
# open amazon 
driver = webdriver.Chrome()
driver.get("https://www.amazon.com")
```

```python
#women clothes -> active wear
input_search = driver.find_element(By.ID, 'twotabsearchtextbox')
search_button = driver.find_element(By.XPATH, "(//input[@type='submit'])[1]")
input_search.send_keys("women active wear")
sleep(1)
search_button.click()
```

```python
urls=[]
titles=[]
asins = []
manufacturers = []
departments = []
```

```python
#getting urls for women active wear

for i in range(6,10):
    print("i am on page no. ", i+1)
    url_on_main_page = driver.find_elements(By.XPATH , "//a[@class='a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal']")
   
    for title in url_on_main_page:
        titles.append(title.text)
    for u in url_on_main_page:
        url = u.get_attribute('href')
        if url:
            urls.append(url)


    next_button = driver.find_element(By.XPATH, "//a[text()='Next']")
    next_button.click()
    sleep(2)
print(urls)    
    


```

```python

asins = []
manufacturers = []
departments = []

for url in urls:
    driver.get(url)
    sleep(3)
    try: 
        details_div = driver.find_element(By.ID, 'detailBullets_feature_div')

        ul_element = details_div.find_element(By.CLASS_NAME, 'a-unordered-list')

        li_elements = ul_element.find_elements(By.TAG_NAME, 'li')

        for li in li_elements:
            text = li.text.strip()
            
            if text.startswith("ASIN"):
                asins.append(text)
                print("asin: ",text)
            elif text.startswith("Manufacturer"):
                manufacturers.append(text)
                print("manufacturer: ",text)
            elif text.startswith("Department"):
                departments.append(text)
                print("department: ",text)
    except Exception as e:
        print("error. different format for details")            
# print("ASINs:", asins)
# print("Manufacturers:", manufacturers)
# print("Departments:", departments)
```

```python
#about this item -> description:
descriptions = []
category = []
subcategory = []
counter = 0
a = "women fashion"
b = "dresses,activewear,leggings,tshirts"
for url in urls[:400]:
    driver.get(url) 
    sleep(3) 
    try:  
        parent_div = driver.find_element(By.ID, 'productFactsDesktopExpander')
        child_div = parent_div.find_element(By.CLASS_NAME, 'a-expander-content')
        ul_elements = child_div.find_elements(By.TAG_NAME, 'ul')
        counter+=1
        print("count = ",counter ,"/n")
        ul_texts = []
        for index in range(1, len(ul_elements) + 1):
            ul_element = ul_elements[index - 1]
            ul_texts.append(ul_element.text + " - ")
        description = ''.join(ul_texts)
        descriptions.append(description)
        print("about the item:\n" ,description )
        #----------------------------category-------------------------------
        category.append(a)        
        print("category: ",a)
        #---------------------------sub category=-----------------------        
        subcategory.append(b)
        print("sub category: ",b)                
        #-----------------------description---------------------------------------
    except Exception as e:
        print("error")

 
#print("description about product :", descriptions)
```

```python
#ratings
ratings = []
for url in urls[:400]:
    driver.get(url)
    sleep(3)
    try:
        r1 = driver.find_element(By.XPATH, '//*[@id="acrPopover"]')
        r2 = r1.get_attribute('title')
        ans = float(r2.split()[0]) 
        print("rating of product: ",ans)
        ratings.append(ans)
    except Exception as e:
        print("error")
#print(ratings)    
```

```python
#brand name from top
brand_original = []
for url in urls[:400]:
    driver.get(url)
    sleep(3)
    try:
        div = driver.find_element(By.ID, 'titleBlockLeftSection')
        tt = div.find_element(By.ID, 'bylineInfo')
        name =tt.text
        name=name.replace("Visit the ","").strip()
        brand_original.append(name)
        print("brand name: ",name)
    except Exception as e:
        print("error")
#print(brand_original)    
```

```python
# for image and colour:
url_images_colors = {}
for i, url in enumerate(urls[:400], start=0):
    print(url)
    sleep(3)
    driver.get(url)
    sleep(3)  
    try:
        color = driver.find_element(By.ID, 'variation_color_name')
        uls = color.find_element(By.CLASS_NAME, 'a-unordered-list')
        liz = uls.find_elements(By.TAG_NAME, 'li')
        imglist = []
        clrlist = []
        for li in liz:
            img_element = li.find_element(By.TAG_NAME, 'img')
            src = img_element.get_attribute('src')
            alt = img_element.get_attribute('alt')
            imglist.append(src)
            clrlist.append(alt)
        
        url_images_colors[i] = {'src': imglist, 'alt': clrlist}
        print("images and colours:\n " , url_images_colors[i])
        
    except Exception as e:
        print("error")
    
# for url, images in url_images_colors.items():
#     print("url", url , "\n")
#     print("image links:", images['src'])
#     print("colours:", images['alt'])
```

```python
#price:
pprice = []
for url in urls[:400]:
    driver.get(url)
    sleep(3)  
    try:
        div = driver.find_element(By.CSS_SELECTOR, '.a-price')
        pp = div.text.split('\n')[0] #because when i scrap it , it gives \n with price
        pprice.append(pp)
        print(pprice)
    except Exception as e:
        print("error")    
```

```python

#images from left side:
url_images = {}
for i, url in enumerate(urls[:400], start=0):
    print(url)
    sleep(3)
    driver.get(url)
    sleep(3)  
    try:
        imggg = driver.find_element(By.ID, 'thumbImages')
        imges=imggg.find_element(By.XPATH , ".//ul[contains(@class, 'a-unordered-list')]")
        imggs=imges.find_elements(By.TAG_NAME, 'img')
        src_list = []
        
        for img in imggs:
            img_element_src = img.get_attribute('src')
            src_list.append(img_element_src)
        url_images[i] = {'src': src_list}
        print("image on left:\n",url_images[i])
        
    except Exception as e:
        print("Error")
# for url, images in url_images.items():
#     print(f"For URL {url}:")
#     print("Srcs:", images['src'])
#     print("************************************")
```

```python
#reviews
reviews = {}
for i, url in enumerate(urls[:400], start=0):
    print(url)
    sleep(3)
    driver.get(url)
    sleep(3)
    try:
        view_more = driver.find_element(By.XPATH, '//*[@id="reviews-medley-footer"]/div[2]/a')
        view_more.click()
        for page in range(5):
            try:
                css = driver.find_elements(By.CSS_SELECTOR, '.a-section.review')
                print("reviews: \n")
                for ccss in css:
                    r_name = []
                    date_r = []
                    review = {}
                    name = ccss.find_element(By.CLASS_NAME, 'a-profile-name').text
                    ttext = ccss.find_element(By.CLASS_NAME, 'review-text').text
                    date = ccss.find_element(By.CLASS_NAME, 'review-date').text
                    review = {'r_name': name, 'date_r': date, 'review': ttext}
                    reviews.setdefault(i, []).append(review) # we will stpre in a dictionary
                    print("Reviewer:", name, "\nReview Date:" , date, "\nReview Text:" , ttext , "\n")

            except selenium.common.exceptions.NoSuchElementException:
                print("No reviews on this page")
            try:
                next_page_button = driver.find_element(By.XPATH, '//*[@id="cm_cr-pagination_bar"]/ul/li[2]/a')
                next_page_button.click()
                time.sleep(2) 
            except selenium.common.exceptions.NoSuchElementException:
                print("No more reviews")
                break

        print("\n")

    except Exception as e:
        print("Error")

driver.quit()

# # Print reviews
# for url, review_list in reviews.items():
#     print(f"For URL {url}:")
#     for review in review_list:
#         print("Date:", review['date_r'])
#         print("Name:", review['r_name'])
#         print("Review:", review['review'])
#         print("------")
```
